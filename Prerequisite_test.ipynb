```python
# 1) List Operations and Basic Statistics

l=[i for i in range(1,21)]
print(l)

even_sum=0
for j in l:
    if j%2==0:
        even_sum+=j
print("Sum of all even numbers in the list:",even_sum)

product=1
for j in l:
    if j%2!=0:
        product*=j
print("product of all odd numbers in the list:",product)

mean = sum(l)/len(l)
print("Mean:",mean)

sorted_list = sorted(l)
n = len(sorted_list)
if n % 2 == 1:
    median = sorted_list[n // 2]
else:
    median = (sorted_list[n // 2 - 1] + sorted_list[n // 2]) / 2
print("Median:",median)

mean_diff_squared = [(x - mean) ** 2 for x in l]
sd = (sum(mean_diff_squared) / (n-1)) ** 0.5
print("Standard deviation:",round(sd,2))
```

```python
#3)FIBONACCI SEQUENNCE

import matplotlib.pyplot as plt

n=int(input())
a=0
b=1
if n==1:
    l=[0]
else:
    l=[0,1]
    while len(l)<n:
        f=a+b
        l.append(f)
        a=b
        b=f
print(f"The first {n} Fibonacci numbers are: {l}")
X=[i for i in range(1,n+1)]


plt.plot(X,l, marker='o', label='Fibonacci Sequence')
plt.xticks(X)
plt.title("Fibonacci Sequence")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
#7) Dictionary Operations
#Bar chart
import matplotlib.pyplot as plt
import numpy as np


n=int(input("Enter the no of students:" ))
d={}
for _ in range (n):
    student=str(input())
    marks=float(input())
    if student not in d.keys():
        d.update({student:marks})
    else:
        d[student]+=marks
        d.update({student:marks})

lst1=[i for i in d.values()]
average=sum(lst1)/len(lst1)
print("Average marks:",round(average,2))

print("Students scoring above average are:")
for a,b in d.items():
    if b>average:
        print(a)

x=[i for i in d.keys()]
y=[i for i in d.values()]

plt.bar(x,y, label='Marks distribution',width=0.4)
plt.xticks(x,rotation=45)
plt.yticks(np.arange(0, 101, 10))
plt.title("Student Marks Distribution")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 16) Set Operations
#VENN DIAGRAMS

import matplotlib.pyplot as plt
from matplotlib_venn import venn2

l1=eval(input("List1:"))
l2=eval(input("List2:"))

s1=set(l1)
s2=set(l2)

intersec=s1.intersection(s2)
un=s1.union(s2)
sym_dif=s1.symmetric_difference(s2)

print("Intersection:",list(intersec))
print("Union:",list(un))
print("Symmetric difference:",list(sym_dif))

plt.figure(figsize=(6, 6))
venn=venn2([s1, s2],set_labels=('Set 1', 'Set 2'))
plt.title("Venn Diagram of two sets")
plt.show()
```

```python
#18) Word count and frequnecy
import matplotlib.pyplot as plt

f2=input("Enter the file:")
f=open(f2,'r')
d={}
for line in f:
    words=line.strip().lower().split()
    for word in words:
        word = word.strip(",.;!?\"'")
        if word:
            d[word]=d.get(word,0)+1
minn=0
longest_word=''
summ=0
for a,b in d.items():
    summ+=b
    if b>minn:
        minn=b
        longest_word=a
print("No of words in the file:", summ)
print("Most frequent word :",longest_word)

x=[i for i in d.keys()]
y=[i for i in d.values()]

plt.bar(x, y, label='Word Frequency',edgecolor='black')
plt.xticks(x,rotation=90)
plt.legend()
plt.title("Frequency of words in text file")
plt.xlabel('Word')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
#19)Random Number Generation

import matplotlib.pyplot as plt
import random

# Generate 100 random integers between 1 and 50
l = [random.randint(1, 50) for _ in range(100)]
print("Random integers:",l)
f=open('random_int','a')
for i in l:
    f.write(f"{i}\n")
f.close()

summ=0
f=open('random_int','r')
for i in f:
    summ+=int(i)
mean= summ/100
print("Mean:",mean)

plt.hist(l, bins=range(1, max(l) + 2), edgecolor='black')
plt.title("Histogram")
plt.xlabel("Random Integer")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
#20) CSV File Operations

import matplotlib.pyplot as plt

filename = input("Enter your file name in .csv format:")
column = input("Enter column:")

data_column = []

with open(filename, 'r') as file:
    header = file.readline().strip().split(',')
    column_index = header.index(column)

    for line in file:
        values = line.strip().split(',')
        data_column.append(float(values[column_index]))

column_sum = sum(data_column)
print(f"Sum of {column}: {column_sum}")

plt.plot(data_column, label=f'{column} values', marker='o')
plt.title(f'Line plot of {column}')
plt.xlabel('Index')
plt.ylabel(f'{column} Values')
plt.legend()
plt.grid()
plt.show()
```

```python
#21) Basic statstics

import matplotlib.pyplot as plt

data = eval(input("Enter list of numbers:"))

mean = sum(data)/len(data)
print("Mean:",round(mean,2))

sorted_data = sorted(data)
n = len(sorted_data)
if n % 2 == 1:
    median = sorted_data[n // 2]
else:
    median = (sorted_data[n // 2 - 1] + sorted_data[n // 2]) / 2
print("Median:", median)

freq = {}
for i in data:
    freq[i] = freq.get(i, 0) + 1
minn=0
mode=0
for a,b in freq.items():
    if b>minn:
        minn=b
        mode=a
print("Mode:",mode)

mean_diff_squared = [(x - mean) ** 2 for x in data]
sd = (sum(mean_diff_squared) / (n-1)) ** 0.5
print("Standard Deviation:",round(sd,2))

plt.boxplot(data)
plt.title("Box Plot of the Data")
plt.ylabel("Values")
plt.show()

```

```python
#22) Normalization
import matplotlib.pyplot as plt

l=eval(input("List of numbers:"))
minn=min(l)
maxx=max(l)
data=[]
for i in l:
    normalized=((i - minn) / (maxx- minn))
    data.append(round(normalized,2))
print("Original numbers:",l)
print("Normalized numbers:", data)

plt.plot(l, label='Original Data', marker='o')
plt.plot(data, label='Normalized Data', marker='x', color='red')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Normalization of Data (Original vs Normalized)')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
